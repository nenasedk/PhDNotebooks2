```python
import numpy as np
import matplotlib.pyplot as plt
from petitRADTRANS.retrieval import plot_style as ps
```

```python
tag = ''
planetspec_b = np.genfromtxt("observations/distance_normalised_observations/YSES1b.txt").T
ydiamond_b = np.genfromtxt(f"grid_fit_spectra/YSES 1 b_diamondback{tag}_R500_BestFitChi2.dat").T
yexo_b = np.genfromtxt(f"grid_fit_spectra/YSES 1 b_exo-rem{tag}_R500_BestFitChi2.dat").T
atmo_b = np.genfromtxt(f"grid_fit_spectra/YSES 1 b_petrus2023{tag}_R500_BestFitChi2.dat").T

planetspec_c = np.genfromtxt("observations/distance_normalised_observations/YSES1c.txt").T
ydiamond_c = np.genfromtxt(f"grid_fit_spectra/YSES 1 c_diamondback{tag}_R500_BestFitChi2.dat").T
yexo_c = np.genfromtxt(f"grid_fit_spectra/YSES 1 c_exo-rem{tag}_R500_BestFitChi2.dat").T


fig, ax = plt.subplots(figsize = (12,12), nrows = 2, sharex=True)
ax[0].errorbar(planetspec_b[0],planetspec_b[1],yerr=3*planetspec_b[2],marker = None, elinewidth = 1.5, color = 'k', label = "YSES 1 b",linewidth = 0)
ax[0].plot(ydiamond_b[0],ydiamond_b[1], linewidth = 3, label = "Diamondback",color = ps.prt_colours[0],alpha = 0.8)
ax[0].plot(yexo_b[0],yexo_b[1], linewidth = 3, label = "ExoRem",color = ps.prt_colours[1],alpha = 0.8)
ax[0].plot(atmo_b[0],atmo_b[1], linewidth = 3, label = "ATMO",color = ps.prt_colours[3],alpha = 0.8)

ax[1].errorbar(planetspec_c[0],planetspec_c[1],yerr=3*planetspec_c[2],marker = None, elinewidth = 1.5, color = 'k', label = "YSES 1 c",linewidth = 0)
ax[1].plot(ydiamond_c[0],ydiamond_c[1], linewidth = 3,color = ps.prt_colours[0],alpha = 0.8)
ax[1].plot(yexo_c[0],yexo_c[1], linewidth = 3,color = ps.prt_colours[1],alpha = 0.8)

ax[1].set_xlim(4.95,11.05)
ax[0].set_ylim(9.6e-18,2.3e-16)
ax[1].set_ylim(4.6e-19,9.6e-18)

ax[0].set_yscale("log")
ax[1].set_yscale("log")

ax[0].legend()
ax[1].legend()
ax[1].set_xlabel("Wavelength [micron]")
ax[0].set_ylabel(r"F$_{\lambda}$ [W/m2/micron]")
ax[1].set_ylabel(r"F$_{\lambda}$ [W/m2/micron]")

plt.savefig("plots/YSES1_ModelFits_v1.pdf",bbox_inches = 'tight')
```
