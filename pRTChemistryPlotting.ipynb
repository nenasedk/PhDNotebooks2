```python
from petitRADTRANS.physics import guillot_global
pressures = np.logspace(-6,2,100)
inds = np.where(pressures>1e-4)
temperatures = guillot_global(pressures,1e-2,1.0,1e4,1100,50)

CO = 0.7
FEH = 0.5
pm_abunds = pm.interpol_abundances(CO*np.ones(100), FEH*np.ones(100), temperatures, pressures)
del pm_abunds['nabla_ad']
nfracs = mass_to_number(pm_abunds)
rat1 = np.median(nfracs["CO"][inds]/nfracs["CH4"][inds])
print(rat1)

dpm_abunds = pm.interpol_abundances(CO*np.ones(100), FEH*np.ones(100), temperatures, pressures, Pquench_carbon=10)
del dpm_abunds['nabla_ad']
dnfracs = mass_to_number(dpm_abunds)
rat2 = np.median(dnfracs["CO"][inds]/dnfracs["CH4"][inds])
print(rat2)
print(rat2/rat1, np.log10(rat2/rat1))

fig,ax = plt.subplots()
ax.plot(temperatures,pressures, linewidth = 4)
ax.set_ylim(1e2,1e-6)
ax.set_yscale('log')
ax2 = ax.twiny()
ax2.plot(pm_abunds["H2O"],pressures,color = 'k', label = "H2O")
ax2.plot(dpm_abunds["H2O"], pressures, color = 'k', linestyle = '--')
ax2.plot(pm_abunds["CO"],pressures,color = 'blue', label = "CO")
ax2.plot(dpm_abunds["CO"], pressures, color = 'blue', linestyle = '--')
ax2.plot(pm_abunds["CH4"],pressures,color = 'red', label = "CH4")
ax2.plot(dpm_abunds["CH4"], pressures, color = 'red', linestyle = '--')
ax2.set_xscale('log')
ax2.legend()
ax2.set_ylim(1e2,1e-6)
ax.set_ylabel("Pressure [bar]")
ax.set_ylabel("Pressure [bar]")
ax.set_xlabel("Temperature [K]")
ax2.set_xlabel("Mass Fraction")
plt.savefig("plots/PM_DisEq_ChemComp.pdf",bbox_inches='tight')
#np.savez("WASP39b_pRTChemTest_co0.3_feh-0.5_1000K",pm_abunds)
```
